```python
with open('/content/drive/MyDrive/Colab Notebooks/fardataset.txt') as f:
    names = [line.rstrip() for line in f]
    names = [i.strip('@') for i in names]
```

```python
# Initial twitter scrape
c = twint.Config()
c.Username = "SmileyTheFed"
#c.Search = "code"
c.Pandas= True #Enable Pandas integration.
c.Limit= 500
nest_asyncio.apply()
twint.run.Search(c)
```

```python
# Create data frame to save tweets into
def columne_names():
  return twint.output.panda.Tweets_df.columns
def twint_to_pd(columns):
  return twint.output.panda.Tweets_df[columns]

data= twint_to_pd(["tweet","username", "name"])
data.head()
```

```python
i=0
while i<len(names):
  c = twint.Config()
  c.Username = names[i]
  #c.Search = "code"
  c.Pandas= True #Enable Pandas integration.
  c.Limit= 500
  nest_asyncio.apply()
  twint.run.Search(c)
  data_new= twint_to_pd(["tweet","username", "name"])
  data = data.append(data_new)
  i+=1
```

```python
labels = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/labelleddata.csv', encoding = 'unicode_escape', engine ='python')
```

```python
merged_data = pd.merge(labels, data)
merged_data.head
```
